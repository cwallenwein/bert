```python
from torch import nn, optim

class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.feed_forward = nn.Linear(10, 2)

    def forward(self, x):
        x = self.feed_forward(x)
        return

model = Model()
optimizer = optim.AdamW(model.parameters(), lr=1e-4)
```

```python
import seaborn as sns

def get_learning_rates(scheduler, epochs):
    lr_rates = []
    for epoch in range(epochs):
        lr_rates += scheduler.get_last_lr()
        scheduler.step()

    return lr_rates

def draw_learning_rates(lr_rates):
    sns.lineplot(lr_rates)

def draw_schedule(scheduler, epochs):
    lr_rates = get_learning_rates(scheduler, epochs)
    draw_learning_rates(lr_rates)
```

```python
from warmup_stable_decay_scheduler import WarmupStableDecayScheduler

scheduler = WarmupStableDecayScheduler(optimizer=optimizer, lr = 1, total_steps = 200, warmup_steps = 50, decay_steps = 100)

learning_rates = get_learning_rates(scheduler, 250)
```

```python
draw_learning_rates(learning_rates)
```
